<p style="font-family:newtimeroman; color:black; font-size:100%; text-align:left; border-radius:10px 10px;">Exercise-1:  What is the normal body temperature for healthy humans? A random sample of 130 healthy human body temperatures provided by Allen Shoemaker yielded 98.25 degrees and standard deviation 0.73 degrees.
Give a 99% confidence interval for the average body temperature of healthy people.
</p> 

```python
sample_mean = 98.25
standard_deviation = 0.73
sample_size = 130
confidence_level = 0.99

critical_value = stats.norm.ppf((1 + confidence_level) / 2)

margin_of_error = critical_value * (standard_deviation / math.sqrt(sample_size))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

confidence_interval = (lower_bound, upper_bound)

print("Confidence Interval:", confidence_interval)
```

<p style="font-family:newtimeroman; color:black; font-size:100%; text-align:left; border-radius:10px 10px;">Answer-1: We received  99% confidence interval for the average body temperature of healthy people based on the provided sample.</p> 

<p style="font-family:newtimeroman; color:black; font-size:100%; text-align:left; border-radius:10px 10px;"> Exercise-2:
The administrators for a hospital wished to estimate the average number of days required for inpatient treatment of patients between the ages of 25 and 34. A random sample of 500 hospital patients between these ages produced a mean and standard deviation equal to 5.4 and 3.1 days, respectively.
Construct a 95% confidence interval for the mean length of stay for the population of patients from which the sample was drawn.</p> 

```python
sample_mean = 5.4
standard_deviation = 3.1
sample_size = 500
confidence_level = 0.95

critical_value = stats.norm.ppf((1 + confidence_level) / 2)

margin_of_error = critical_value * (standard_deviation / math.sqrt(sample_size))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

confidence_interval = (lower_bound, upper_bound)
print("Confidence Interval:", confidence_interval)
```

<p style="font-family:newtimeroman; color:black; font-size:100%; text-align:left; border-radius:10px 10px;">Answer-2: We received  95% confidence interval for the mean length of stay for the population of patients based on the provided sample.</p> 

<p style="font-family:newtimeroman; color:black; font-size:100%; text-align:left; border-radius:10px 10px;"> Exercise-3:
The hourly wages in a particular industry are normally distributed with mean $13.20 and standard deviation $2.50. A company in this industry employs 40 workers, paying them an average of $12.20 per hour. Can this company be accused of paying substandard wages? Use an α = .01 level test. (Wackerly, Ex.10.18)

CHECK: statistic: -2.5298221281347035, pvalue= 0.005706018193000826</p> 

```python
sample_mean = 12.20
population_mean = 13.20
population_std = 2.50
sample_size = 40
alpha = 0.01

t_statistic = (sample_mean - population_mean) / (population_std / (sample_size ** 0.5))
p_value = stats.t.sf(abs(t_statistic), sample_size - 1) * 2


print("Statistic:", t_statistic,"\np-value:", p_value)
```

<p style="font-family:newtimeroman; color:black; font-size:100%; text-align:left; border-radius:10px 10px;">Answer-3: We reject the null hypothesis because the p-value is less than the significance level.</p>
